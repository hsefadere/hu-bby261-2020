```
#pycharm programında çalıştırınız
from tkinter import *
import string


pencere=Tk()

baslik = pencere.title("sezar")
def alfabe():

    butunalfabe = string.printable
    alfabe = ["ç", "ğ", "ı", "ö", "ş", "ü", "Ç", "Ğ", "İ", "Ö", "Ş", "Ü"]
    for i in butunalfabe:
        alfabe.append(i)
    return alfabe
def sifrele():

    a = alfabe()
    metin = sifrelemek.get()
    anahtar = 2
    sifrelenen =""
    atama = open("Sezar Şifreliler.txt", "w", encoding="utf-8")
    for k in metin:
        sifrelenen += a[(a.index(k) + anahtar) % len(a)]
    atama.write(sifrelenen)
    atama.close()
    ilk.insert(0,sifrelenen)



def sifrecoz():
    a = alfabe()
    anahtar = 2
    sifrecoz = ""
    sezoku = open("Sezar Şifreliler.txt","r",encoding="utf-8")

    for x in sezoku.read():
        sifrecoz += a[(a.index(x) - anahtar)%len(a)]
    ikinci.insert(0,sifrecoz)

def sifrekir():
    alfabem = alfabe()
    metinİcinden = "sorun devlet sahip sıra yüzde ay olay söz sistem kapı kitap gece alan bugün dönem arkadaş ürün aile erkek güç gerçek ilişki çevre yaşam halk sokak bey tarih özellik bölüm akıl anlam banka ayak toplum araç madde tür karar hava sayı grup oda biçim haber Allah soru arka yazı okul dil şirket kaynak program hareket renk hak çalışma açı parça gazete değer yapı doktor gelir görev amaç bölge film müşteri telefon eğitim deniz etki vücut düşünce milyon temel kültür resim ışık hanım hatun hizmet ihtiyaç nokta yön oyun işlem oran orada dikkat bilgisayar gelecek oğul lira üretim dakika araba ağız duygu örnek derece duvar sanat ana hastalık öğrenci televizyon yöntem masa takım kafa müzik enerji üniversite spor türlü can kısım ölüm sağlık sabah internet teknik dışarı merkez ortam düzey köy yönetim aşağı cevap toprak isim akşam araştırma kan hasta şehir hafta trafik hesap otomobil yabancı davranış mutfak kent fiyat kol cam önem koca varlık ilgi satış içeri acı kat ekonomi fotoğraf hayvan savaş mal saç kalan sayfa teknoloji"

    kelimeInList = metinİcinden.split()

    kilit = 0
    sezok = open("Sezar Şifreliler.txt", "r", encoding="utf-8")
    metin = sezok.read()



    while kilit <= len(alfabem):
        kilit += 1
        sifresikirilan = ""
        for hrn in metin:
            sifresikirilan += alfabem[(alfabem.index(hrn) + kilit) % len(alfabem)]

        for j in range(len(kelimeInList)):
            if sifresikirilan.find(kelimeInList[j]) != -1:
                sonuc.insert(0,sifresikirilan)





baslikk = Label(text="Sezar Şifreleme",fg="black",bg="yellow",width="20", font=("Monotype Corsiva", 30))
baslikk.grid(row=1,column=1)

sifrelemek = Entry(width=40)
sifrelemek.grid(row=2,column=1)

ilk = Entry(width=40)
ilk.grid(row=3,column=0)

etiket1 = Label(text="Şifreli Cümle",fg="white",bg="#0f14dd",font=("Monotype Corsiva", 22))
etiket1.grid(row=4,column=0)

buton1 = Button(text="Şifrele",font=("cabria", 13), fg="red",command= lambda:sifrele())
buton1.grid(row=5,column=0)

ikinci = Entry(width=40)
ikinci.grid(row=3,column=3)

etiket2 = Label(text="Çözülmüş Şifre",fg="white",bg="#0f14dd",font=("Monotype Corsiva", 22))
etiket2.grid(row=4,column=3)

buton2 = Button(text="Şifre Çöz",font=("cabria", 13), fg="red",command=lambda :sifrecoz())
buton2.grid(row=5,column=3)

etiket4 = Label(text="sifreli cümle",fg="white",bg="#0f14dd",font=("Monotype Corsiva", 22))
etiket4.grid(row=5,column=1)

sonucb= Button(text="Şifre Kır",font=("cabria", 13), fg="red",command=lambda :sifrekir())
sonucb.grid(row=6,column=1)

sonuc = Entry(width=40)
sonuc.grid(row=7,column=1)



mainloop()
```
